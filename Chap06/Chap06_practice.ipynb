```python
import thinkstats2
import math
import numpy as np
```

```python
import thinkplot
```

```python
%matplotlib inline
```

```python
mean, var = 163, 52.8
```

```python
std = math.sqrt(var)
pdf = thinkstats2.NormalPdf(mean, std)
```

```python
pdf.Density(mean+std)
```

```python
thinkplot.Pdf(pdf, label='normal')
thinkplot.Show()
```
